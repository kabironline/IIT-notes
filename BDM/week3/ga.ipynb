```python
import pandas
```

```python
dataset = pandas.read_csv('data.csv')
dataset
```

```python
# What percentage of sample households have only one two wheeler ? (2DF)
q1 = dataset["TWO_WHEELERS_OWNED"].where(dataset["TWO_WHEELERS_OWNED"] == 1).count() / dataset["TWO_WHEELERS_OWNED"].count() * 100
q1 # need to check because do we need to consider the households with no two wheelers
```

```python
# What percentage of sample households don't have a two wheeler, but are interested in buying one?
intereted = dataset.loc[dataset["WILL_BUY_TWO_WHEELER"] == "Y"].count().iloc[0]
no_two_wheeler = dataset.loc[dataset["TWO_WHEELERS_OWNED"] == 0].count().iloc[0]
q2 = intereted / no_two_wheeler * 100
q2
```

```python
# What percentage of sample households already have a two wheeler, but are interested in buying one more?
two_wheeler = dataset.loc[dataset["TWO_WHEELERS_OWNED"] > 0].count().iloc[0]
interested = dataset.loc[dataset["WILL_BUY_TWO_WHEELER_NOW"] == "Y"].count().iloc[0]
q3 = interested / two_wheeler * 100
q3
```

```python
# What percentage of urban households don't have a two wheeler, but are intending to buy immediately?
urban = dataset.loc[dataset["REGION_TYPE"] == "URBAN"].count().iloc[0]
immediate = dataset.loc[dataset["REGION_TYPE"] == "URBAN"].where(dataset["WILL_BUY_TWO_WHEELER_NOW"] == "Y").count().iloc[0]
q4 = immediate / urban * 100
q4
```
