```python
import numpy as np
import pandas as pd
import seaborn as sns
from fitter import Fitter, get_common_distributions, get_distributions
```

### Loading the Dataset

```python
dataset = pd.read_csv('Automobile.csv')
dataset.head()
data = "mpg"
```

- Checking info on the data

```python
dataset.info()
```

### Plotting the dataset

```python
sns.set_style('white')
sns.set_context("paper", font_scale = 2)
sns.displot(data=dataset, x=data, kind="hist", bins = 7, aspect = 1.5)
```

- Extracting the data we need

```python
hp = dataset[data].values
```

### Checking in which distribution the data fits

```python
f = Fitter(hp,
           timeout=6,
           bins=8,
           distributions=['gamma',
                          "beta",
                          "chi2",])
f.fit()
f.summary()
f.get_best(method = "sumsquare_error")
```
